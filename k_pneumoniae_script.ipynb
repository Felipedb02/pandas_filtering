```python
import numpy as np
import pandas as pd
```

```python
datos = pd.read_csv("combined_summary_k_pneumoniae.csv", sep=";", index_col=0)
datos
```

```python
datos=datos.rename_axis('samples')
datos
```

```python
#####------------------>>>>>>> mutaciones de genes

df_assembled = datos[datos.columns.drop(list(datos.filter(regex='.assembled')))]
df_ref_seq = df_assembled[df_assembled.columns.drop(list(df_assembled.filter(regex='.ref_seq')))]
df_pct_id = df_ref_seq[df_ref_seq.columns.drop(list(df_ref_seq.filter(regex='.pct_id')))]
mutations = df_pct_id[df_pct_id.columns.drop(list(df_pct_id.filter(regex='.ctg_cov')))]
mutations
```

```python
mutations_yes=mutations[mutations=="yes"]
mutations_yes
```

```python
mutations_st=mutations_yes.stack()
mutations_st 
```

```python
#####------------------->>>>>>>> mutation dataframe final result
mutations_yes=pd.DataFrame(mutations_st)
mutation_result_df=mutations_yes.rename(columns = {0:'mutation_result'}, inplace = False)
mutation_result_df
```

```python
mutation_result_df = mutation_result_df.reset_index(level=[0,1])
mutation_result_df
```

```python
mutation_result_df=mutation_result_df.stack().str.replace('.','_').unstack()
```

```python
mutation_result_df=mutation_result_df.set_index(['samples', 'level_1'])
mutation_result_df
```

```python
####----------------------------------->>> .assembled process

assembled=datos.columns.str.find(".assembled")
```

```python
AS=assembled != -1
AS
```

```python
asd=datos.iloc[:,AS]
asd
```

```python
yes=asd[asd=="yes"]
yes
```

```python
yes_nonunique=asd[asd=="yes_nonunique"]
yes_nonunique
```

```python
yeah= pd.concat([yes, yes_nonunique], axis=1)
yeah
```

```python
yeah_column=yeah.stack()
yeah_column    
```

```python
yeah_column.index
```

```python
yeah_column_df = yeah_column.reset_index(level=[0,1])
yeah_column_df
```

```python
fixed_level1=yeah_column_df.level_1.str.rstrip('.assembled')
fixed_level2=pd.DataFrame(fixed_level1)
fixed_level2
```

```python
yeah_column_df['level_1']=fixed_level2['level_1']
yeah_column_df
```

```python
#####----------------------------------> data frame 1 ---> assembled arreglados!!
yeah_column_arreglada=yeah_column_df.set_index(['samples', 'level_1'])
yeah_column_arreglada=yeah_column_arreglada.rename(columns = {0:'yes/yes_nonunique'}, inplace = False)
yeah_column_arreglada
```

```python
##### agroupment of the ".ctq_cov" data 
gene_id=datos.columns.str.find(".pct_id")
gene_id
```

```python
ID=gene_id != -1
ID
```

```python
gene_identification=datos.iloc[:,ID]
gene_identification
```

```python
gene_identification_df=gene_identification.stack()
gene_identification_df
```

```python
gene_identification_df = gene_identification_df.reset_index(level=[0,1])
gene_identification_df
```

```python
fixed_ID_level1_df=gene_identification_df.level_1.str.rstrip('.pct_id')
fixed_ID_level1_df=pd.DataFrame(fixed_ID_level1_df)
fixed_ID_level1_df
```

```python
gene_identification_df['level_1']=fixed_ID_level1_df['level_1']
gene_identification_df
```

```python
#####----------------------------------> data frame 2 ---> identificacion de genes arreglados!!
gene_identification_fixed=gene_identification_df.set_index(['samples', 'level_1'])
gene_identification_fixed=gene_identification_fixed.rename(columns = {0:'%_ID'}, inplace = False)
gene_identification_fixed
```

```python
##### ------------------------------------------------>  agroupment of the ".ctq_cov" data 
coverage=datos.columns.str.find(".ctg_cov")
coverage
```

```python
CO=coverage != -1
CO
```

```python
cover=datos.iloc[:,CO]
cover ###----> it shows the locus that have .ctg_cov suffix with a value (if it has a value on it)
```

```python
cover_filtrated=cover[cover>10]  ###----> sort the values coverage higher than 10 
cover_filtrated
```

```python
final_coverage=cover_filtrated.stack()
final_coverage
```

```python
final_coverage_df = final_coverage.reset_index(level=[0,1])
final_coverage_df
```

```python
fixed_level1_cover=final_coverage_df.level_1.str.rstrip('.ctg_cov')
fixed_level1_cover=pd.DataFrame(fixed_level1_cover)
fixed_level1_cover
```

```python
final_coverage_df['level_1']=fixed_level1_cover['level_1']
final_coverage_df
```

```python
#####----------------------------------> data frame 2 ---> coverage arreglados!!
final_coverage_fixed=final_coverage_df.set_index(['samples', 'level_1'])
final_coverage_fixed=final_coverage_fixed.rename(columns = {0:'coverage'}, inplace = False)
final_coverage_fixed
```

```python
database_anthony=pd.read_csv("/Users/felipedb/Documents/data/projects/training/taller_agrosavia/ariba_exercise/exercise_2_ncbi_acquired_genes_metadata.csv", sep=",")
database_anthony
```

```python
NCBI_filtered = database_anthony[database_anthony["refseq nucleotide"].str.find("NG_")!=-1]
NCBI_filtered
```

```python
RF=ref_seq != -1
RF
```

```python
referenced=datos.iloc[:,RF]
referenced
```

```python
NG=referenced.stack()
NG
```

```python
NG_arreglados=pd.DataFrame(NG_separados,index=NG.index)
NG_arreglados.columns=["refseq_nucleotide_accession"]
NG_arreglados
```

```python
NG_arreglados_df = NG_arreglados.reset_index(level=[0,1])
NG_arreglados_df
```

```python
fixed_column_NG_arreglados_df=NG_arreglados_df.level_1.str.rstrip('.ref_seq')
fixed_column_NG_arreglados_df=pd.DataFrame(fixed_column_NG_arreglados_df)
fixed_column_NG_arreglados_df
```

```python
NG_arreglados_df['level_1']=fixed_column_NG_arreglados_df['level_1']
NG_arreglados_df
```

```python
#####----------------------------------> data frame 3 ---> NG arreglados!!
NG_fixed=NG_arreglados_df.set_index(['samples', 'level_1'])
NG_fixed
```

```python
resultado_final=NG_fixed.join(tabla_final_NCBI.set_index("refseq nucleotide"),on="refseq_nucleotide_accession")
resultado_final
```

```python
resultado_integrated= pd.concat([resultado_final, yeah_column_arreglada], axis=1)
resultado_integrated=pd.concat([resultado_integrated, final_coverage_fixed], axis=1)
resultado_integrated=pd.concat([resultado_integrated, gene_identification_fixed], axis=1)
resultado_integrated=pd.concat([resultado_integrated, mutation_result_df], axis=1)
resultado_integrated
```
